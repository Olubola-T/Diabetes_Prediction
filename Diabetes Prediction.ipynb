```python
#data source: https://www.kaggle.com/datasets/slmsshk/medical-students-dataset

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r'medical_students_dataset.csv')
print(df.head())
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.hist(bins=15, figsize=(16,8))
```

```python
df['Temperature'].fillna(value=df['Temperature'].mean(), inplace=True)
df['Temperature'].isna().sum()
```

```python
for column in ['Age', 'Height', 'Weight', 'BMI', 'Heart Rate', 'Blood Pressure', 'Cholesterol']:
    df[column].fillna(value= df[column].median(), inplace= True)
    
df.isna().sum()
```

```python
for column in ['Gender', 'Blood Type', 'Diabetes', 'Smoking']:
    df[column].fillna(method='ffill', inplace= True)

df.isna().sum()
```

```python
df['Smoking'].fillna(method='bfill', inplace= True)
```

```python
df['Student ID']= range(1, 200001)
df.isna().sum()
```

```python
for column in ['Age', 'Heart Rate', 'Blood Pressure', 'Cholesterol']:
    for i in df[column]:
        i = round(i, 0)

print(df.head())
```

```python
df.info()
```

```python
numerical_vars= ['Age', 'Height', 'Weight', 'BMI', 'Temperature', 'Heart Rate', 
                 'Blood Pressure', 'Cholesterol']

categorical_vars= ['Gender', 'Blood Type', 'Diabetes', 'Smoking']
```

```python
plt.figure(figsize=(14,10), layout='constrained')
for i, var in enumerate(numerical_vars):
    plt.subplot(4, 2, i+1)
    sns.boxplot(data=df, x=var)
```

```python
plt.figure(figsize=(6,6), layout='constrained')
for i, var in enumerate(categorical_vars):
    plt.subplot(2, 2, i+1)
    sns.countplot(data=df, x=var)
```

```python
def BMI_conditions(x):
    if x >= 18.5 and x <= 24.9:
        return 'Normal'
    elif x > 24.9 and x <= 30:
        return 'Overweight'
    elif x < 18.5:
        return 'Underweight'
    else:
        return 'Obese'
```

```python
df['BMI Group']= df['BMI'].apply(BMI_conditions)
    
print(df.head())
```

```python
df['Cholesterol Group']= df['Cholesterol'].apply(lambda x: 'Normal' if x <= 200 else 'Abnormal')

print(df.head())
```

```python
hues= ['Smoking', 'Gender', 'Blood Type', 'BMI Group', 'Cholesterol Group']

plt.figure(figsize=(6,10), layout='constrained')
for i, var in enumerate(hues):
    plt.subplot(3, 2, i+1)
    sns.countplot(data=df, x='Diabetes', hue=var)
```

```python
from scipy import stats

# df= df[(np.abs(stats.zscore(df)) < 3).all(axis=1)] can be used if the whole dataframe has numeric variables

df= df[(np.abs(stats.zscore(df['Temperature'])) < 3)]
```

```python
plt.figure(figsize=(14,10), layout='constrained')
for i, var in enumerate(numerical_vars):
    plt.subplot(4, 2, i+1)
    sns.boxplot(data=df, x=var)
```

```python
Q3= df['BMI'].quantile(0.75)
Q1= df['BMI'].quantile(0.25)
IQR= Q3-Q1

upper= Q3 + 1.5*IQR
lower= Q1 - 1.5*IQR

df= df[(df['BMI'] > lower) & (df['BMI'] < upper)]
```

```python
plt.figure(figsize=(14,10), layout='constrained')
for i, var in enumerate(numerical_vars):
    plt.subplot(4, 2, i+1)
    sns.boxplot(data=df, x=var)
```

```python
df.shape
```

```python
x= df[['Age', 'Gender', 'Height', 'Weight', 'Blood Type', 
       'Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol', 'Smoking']]

y= df[['Diabetes']]
```

```python
x.head()
```

```python
y.head()
```

```python
encode_col= ['Gender', 'Blood Type', 'Smoking']

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in encode_col:
    x[col]= le.fit_transform(x[col])
    
x.head()
```

```python
le = LabelEncoder()
y= le.fit_transform(y)
print(y[0:5])
```

```python
print(x.var())
```

```python
normalize = ['Age', 'Height', 'Weight', 'Heart Rate', 'Blood Pressure', 'Cholesterol']

for col in normalize:
    x[col]= np.log(x[col])
    
x.head()
```

```python
x.var()
```

```python
df['Diabetes'].value_counts()
```

```python
from sklearn.model_selection import train_test_split, KFold, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.3, stratify=y, random_state= 17)
```

## Logistic Regression

```python
lr= LogisticRegression()

lr.fit(x_train, y_train.ravel())
```

```python
y_pred_lr= lr.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_lr)
```

```python
confusion_matrix(y_test, y_pred_lr)
```

```python
print(classification_report(y_test, y_pred_lr))
```

## Decision tree classifier

```python
dt= DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train.ravel())

y_pred_dt= dt.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_dt)
```

```python
print(confusion_matrix(y_test, y_pred_dt))
```

## K neighbours classifier

```python
kn= KNeighborsClassifier()
kn.fit(x_train, y_train.ravel())
y_pred_kn = kn.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_kn)
```

```python
print(confusion_matrix(y_test, y_pred_kn))
```

## Support vector classifier

```python
sv= SVC()
sv.fit(x_train, y_train.ravel())
y_pred_sv = sv.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_sv)
```

```python
print(confusion_matrix(y_test, y_pred_sv))
```

## Hyperparameter tuning

```python
from sklearn.model_selection import KFold, RandomizedSearchCV
```

```python
kf= KFold(n_splits=5, shuffle=True, random_state=17)
```

```python
lr_params = {"tol": np.linspace(0.0001, 1.0, 50),
         "C": np.linspace(0.1, 1.0, 50),
         "class_weight": ["balanced", {0:0.8, 1:0.2}],
             "solver": ['lbfgs', 'liblinear', 'newton-cholesky', 'saga']}

# Instantiate the RandomizedSearchCV object
lr_cv = RandomizedSearchCV(lr, lr_params, cv=kf)

# Fit the data to the model
lr_cv.fit(x_train, y_train.ravel())
```

```python
print(lr_cv.best_score_, lr_cv.best_params_)
```

```python
y_pred_lr_cv= lr_cv.predict(x_test)
accuracy_score(y_test, y_pred_lr_cv)
```

```python
print(confusion_matrix(y_test, y_pred_lr_cv))
```

```python
print(classification_report(y_test, y_pred_lr_cv))
```

```python
kn_params = {"n_neighbors": range(1, 11),
         "leaf_size": range(10, 60, 10),
         "algorithm": ['auto', 'ball_tree', 'kd_tree', 'brute'],
         "weights": ['uniform', 'distance']}

# Instantiate the RandomizedSearchCV object
kn_cv = RandomizedSearchCV(kn, kn_params, cv=kf)

# Fit the data to the model
kn_cv.fit(x_train, y_train.ravel())
```

```python
print(kn_cv.best_score_, kn_cv.best_params_)
```

```python
y_pred_kn_cv= kn_cv.predict(x_test)
accuracy_score(y_test, y_pred_kn_cv)
```

```python
print(confusion_matrix(y_test, y_pred_kn_cv))
```

```python
print(classification_report(y_test, y_pred_kn_cv))
```
